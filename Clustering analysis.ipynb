# # For Features missing values less than 30% we use knn method

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder
```

```python
Country=pd.DataFrame(df['Country'])
Country
```

```python
df
```

- For missing values less than 25% with skewed data and outliers, you may want to consider using the robust imputation method. This method is designed to be less sensitive to outliers and can work well with skewed data.

- For missing values more than 50%, you may want to consider using the multiple imputation method. This method generates multiple imputed datasets and takes into account the uncertainty of the imputations when analyzing the data. This can be useful for handling missing values in the presence of a large amount of missing data

- For missing values less than 25%, you can use simpler imputation methods like k-NN or mean imputation.

- For missing values more than 50%, you can use multiple imputation methods to generate multiple imputed datasets.

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from fancyimpute import IterativeImputer

# Impute missing values less than 25% using mean imputation
imputer1 = SimpleImputer(strategy='median')
df1_imputed = imputer1.fit_transform(df.loc[:, df.isna().sum()/len(df) <= 0.25])
df1 = pd.DataFrame(df1_imputed, columns=df.loc[:, df.isna().sum()/len(df) <= 0.25].columns)

# Impute missing values between 25-50% using iterative imputation
imputer2 = IterativeImputer(max_iter=10, random_state=0)
df2_imputed = imputer2.fit_transform(df.loc[:, (df.isna().sum()/len(df) > 0.25)])
df2 = pd.DataFrame(df2_imputed, columns=df.loc[:, df.isna().sum()/len(df) > 0.25].columns)

# Combine the imputed dataframes
df_imputed1 = pd.concat([df1, df2], axis=1)
```
